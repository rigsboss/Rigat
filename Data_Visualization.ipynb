# Understanding the Data Analysis Workflow with a Focus on Data Visualization

# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
df = pd.read_csv("Mall_Customers.csv")
```

# Display basic dataset information

```python
print(df.info())  # Shows column names, data types, and missing values
print(df.head())  # Displays the first few rows of the dataset
```

# Distribution of customer spending

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Spending Score (1-100)"], bins=30, kde=True)
plt.title("Distribution of Customer Spending")
plt.xlabel("Total Spending ($)")
plt.ylabel("Number of Customers")
plt.show()
```

# What This Does:

Loads the dataset.

Displays basic dataset information (columns, data types, missing values).

Plots a histogram to show the distribution of customer spending.


# Relationship between Age and Spending

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Age"], y=df["Spending Score (1-100)"], hue=df["Gender"])
plt.title("Customer Age vs. Spending")
plt.xlabel("Age")
plt.ylabel("Total Spending ($)")
plt.legend(title="Segment")
plt.show()
```

#What This Does:

Uses a scatter plot to visualize the relationship between customer Age and Total Spending.

Colors the points based on Customer Segment, making it easier to spot trends.

# Spending by Customer Segment


```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Annual Income (k$)", y="Spending Score (1-100)", data=df)
plt.title("Spending Distribution by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Total Spending ($)")
plt.show()
```

#What This Does:

Uses a box plot to show the spending distribution across different customer segments.

Helps identify which segment spends the most and whether there are outliers.

#Key Takeaways:

The histogram helps understand overall spending behavior.

The scatter plot shows if age influences spending.

The box plot reveals spending patterns across customer segments.

#Insights from the Box Plot:

1. Outliers  Outliers appear as individual dots outside the "whiskers" of the box plot. These are customers whose spending is significantly different from the rest of their segment.

- Outliers above the whiskers: High spenders (possibly VIP or impulsive buyers).

- Outliers below the whiskers: Extremely low spenders, potentially inactive customers.

2. Segment Comparisons  You can compare which customer segment spends the most on average and which has the widest range of spending behaviors.

#Next Steps Based on the Insights:

- High-spending outliers might represent loyal or premium customers; consider targeting them with exclusive offers.

- Low-spending outliers may need targeted promotions to increase engagement.

- If one segment has low median spending and a narrow range, it may not be a profitable target for marketing efforts.
